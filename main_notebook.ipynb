```python
import os
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv(os.path.join('data','train.csv')).dropna()
train_df.head()
```

```python
test_df = pd.read_csv(os.path.join('data','test.csv')).dropna()
test_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

home_planets = train_df.HomePlanet.unique()
home_planets_encoded = LabelEncoder().fit(home_planets)

cabin = train_df.Cabin.unique()
cabin_encoded = LabelEncoder().fit(cabin)

destination = train_df.Destination.unique()
destination_encoded =  LabelEncoder().fit(destination)

train_df['CryoSleep'] = train_df.CryoSleep.apply(lambda x: 1 if x == 'True' else 0)
train_df['VIP'] = train_df.VIP.apply(lambda x: 1 if x == 'True' else 0)
train_df['HomePlanet'] = home_planets_encoded.transform(train_df['HomePlanet'])
train_df['Cabin'] = cabin_encoded.transform(train_df['Cabin'])
train_df['Destination'] = destination_encoded.transform(train_df['Destination'])

X = train_df.drop(['PassengerId','Transported','Name'], axis=1)

y = train_df.Transported.apply(lambda x: 1 if x == True else 0).values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Scikit-Learn

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
clf.fit(X_train, y_train.ravel())
clf.score(X_test, y_test.ravel())
```
